# Different ways Of Creating Series

````python
import pandas as pd
````

````python
#Creating an Empty Series
s=pd.Series()
s
````

````python
#Creating Series with list
l=[1,2,3,4,5]
s=pd.Series(l)
s
````

````python
#we can also change the index of series by
l=[1,2,3,4,5]
s=pd.Series(l,index=['a','b','c','d','e'])
s
````

````python
#Creating Series with Array(space seperated values in list is array)
import numpy as np
a=np.array([80,20,78])
s=pd.Series(a)
s
````

````python
#Creating Series with Dictionary
d={'name':'miraj','roll':10,'age':15}
s=pd.Series(d)
s
````

# Important Attibutes for Series and Data Frame

index - series.index - returns all index values

array - series.array - returns all data values in pandas array (work only for series)

dtype - series.dtype - returns the datatype of the values (works only for series)

values - series.values - returns values of series in numpy array

shape - series.shape - returns the shape 

ndim - series.ndim - returns the dimension of the series

size - series.size - returns the size of the series

empty- series.empty - returns true if series is empty and false if series is not empty

# To concat dataframes

````python
df1=pd.DataFrame({'Name':['miraj','aman'],'Roll':[10,11]})
df2=pd.DataFrame({'Name':['arbit','amir'],'Roll':[15,13]})
df3=pd.DataFrame({'Name':['june','july'],'Roll':[18,21]})

df=pd.concat([df1,df2,df3],keys=['group1','group2','group3'],names=['group','rowno'])
df
````
